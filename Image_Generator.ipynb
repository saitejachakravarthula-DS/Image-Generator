```python
from IPython.display import display
import ipywidgets as widgets
from PIL import Image
import torch
from torch.amp import autocast
from diffusers import StableDiffusionPipeline
```

```python
# Hugging Face Authentication Token (replace with your token)
auth_token = "Replace with your Hugging Face token"
```

```python
# Load the Stable Diffusion model
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"

print("Loading the Stable Diffusion model...")
pipe = StableDiffusionPipeline.from_pretrained(
    model_id, revision="fp16", torch_dtype=torch.float16, use_auth_token=auth_token
)
pipe.to(device)
print("Model loaded successfully!")
```

```python
# Widgets for the user interface
prompt_input = widgets.Text(
    value='A scenic mountain landscape',
    placeholder='Enter your prompt here',
    description='Prompt:',
    layout=widgets.Layout(width="100%")
)

generate_button = widgets.Button(description="Generate Image")
output = widgets.Output()
```

```python
# Callback function to generate the image
def generate_image(_):
    with output:
        output.clear_output()
        if device == "cpu":
            print("Warning: CUDA not available, generation will be slow.")
        display(widgets.Label("Generating..."))
        prompt = prompt_input.value
        with autocast("cuda"):
            image = pipe(prompt, guidance_scale=8.5).images[0]
        image.save("generated_image.png")
        display(image)

generate_button.on_click(generate_image)
```

```python
# Display the interface
display(prompt_input, generate_button, output)
```
